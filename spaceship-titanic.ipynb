```python
# 결측치를 처리합니다.
# HomePlanet과 Destination은 최빈값으로 대체했으며, 여부를 확인하는 CyroSleep과 VIP는 Boolean으로 변환
# Age는 중앙값으로 대체하였고, Name은 Unknown으로 대체하였음

pd.set_option('future.no_silent_downcasting', True)
data['HomePlanet'] = data['HomePlanet'].fillna(data['HomePlanet'].mode()[0])
data['Destination'] = data['Destination'].fillna(data['Destination'].mode()[0])
data['CryoSleep'] = data['CryoSleep'].fillna(False)
data['VIP'] = data['VIP'].fillna(False)
data['Age'] = data['Age'].fillna(data['Age'].median())
amenities = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
for col in amenities:
    data[col] = data[col].fillna(0)
data['Name'] = data['Name'].fillna('Unknown')
data['Cabin'] = data['Cabin'].fillna('Unknown')

print(data.isnull().sum())
```

```python
transported_counts = data['Transported'].value_counts().reset_index()
transported_counts.columns = ['Transported', 'count']
transported_counts['Percentage'] = (transported_counts['count'] / transported_counts['count'].sum()) * 100

transported_counts = pd.DataFrame({
    "Transported": transported_counts['Transported'],
    "count": transported_counts['count'],
    "Percentage(%)": transported_counts['Percentage'].round(2)
}).sort_values(by='Percentage(%)', ascending=False).reset_index(drop=True)

transported_counts
```

```python
import requests
from matplotlib import rc
import matplotlib.font_manager as fm

def set_font():
    url = "https://github.com/google/fonts/raw/main/ofl/merriweather/Merriweather-Regular.ttf"
    font_path = "./data/working/Merriweather-Regular.ttf"
    with open(font_path, "wb") as f:
        f.write(requests.get(url).content)
    font_prop = fm.FontProperties(fname=font_path)
    rc('font', family=font_prop.get_name())
    return font_prop


font_prop = set_font()
```
